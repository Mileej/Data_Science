```python
#importar librerias 
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import make_column_transformer, make_column_selector
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn import set_config
set_config(display='diagram')
```

```python
df=pd.read_csv('/content/drive/MyDrive/Semestre 2022-2/CODING_DOJO_DS/Machine_Learning/sales_predictions.csv')
df.head(10)
```

```python
df.info()
```

```python
#count null
df.isna().any()
```

```python
#eliminamos la columna  Item_Identifier, no es relevante
df=df.drop(['Item_Identifier'], axis=1)
df
```

objetivo (X) y  características (y): columna "Item_Outlet_Sales" como el objetivo y el resto de las variables relevantes como el matriz de características. 

```python
#Definir la característica (X) y objetivo (y).
X=df.drop('Item_Outlet_Sales', axis=1)
y=df['Item_Outlet_Sales']
#train test split en los datos para prepararlos para aprendizaje automático.
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

pipeline de preprocesamiento para preparar el conjunto de datos para el aprendizaje automático

```python
#Identificar cada característica como numérica, ordinal o nominal. (Entreguen esta pregunta en una celda de texto en un cuaderno de Colab).
cat_selector = make_column_selector(dtype_include='object')
num_selector = make_column_selector(dtype_include='number')
```

```python
#Selección de las columnas númericas de cada tipo 
num_columns=num_selector(X_train)
cat_columns=cat_selector(X_train)
```

```python
# observación
print('columnas númericas',num_columns)
print('columnas categoricas', cat_columns) #Tenemos más variables categoricas que númericas
```

```python
#Imputar algún valor faltante. estrategia "mean" para las columnas numéricas y la estrategia "most_frequent" para las columnas categóricas.

#Antes de la imputación aislemos las columnas numericas a las que les faltan datos
df_num =df[num_columns]
#aislamos las columnas con datos faltantes
df_num.loc[:,df_num.isna().any()]
```

```python
X_train.isna().any()
```

```python
#usaremos ColumnTransformer para imputar los valores faltantes tanto númericos como categóricos
mean_imputer= SimpleImputer(strategy='mean')
freq_imputer= SimpleImputer(strategy= 'most_frequent')
```

```python
# codificación ONE-HOT a las características nominales.
ohe=OneHotEncoder(handle_unknown='ignore', sparse=False)

#Escalar las columnas numéricas.
scaler= StandardScaler()
```

```python
#instanciamos pipelines 
numeric_pipe =make_pipeline(mean_imputer, scaler)
numeric_pipe
```

```python
categorical_pipe=make_pipeline(freq_imputer, ohe)
categorical_pipe
```

```python
#Añadimos a un solo objeto
# tuplas para llevar al column transformer
number_tuple=(numeric_pipe, num_selector)
category_tuple=(categorical_pipe, cat_selector)

# unico ColumnTransformer
preprocessor= make_column_transformer(number_tuple, category_tuple)
preprocessor
```

```python
#fit on train 
preprocessor.fit(X_train) #vemos 4 transformadores dentro de un ColumnTransformer
```

```python
#transformar los conjuntos de datos de entrenamiento y prueba
X_train_processed= preprocessor.transform(X_train)
X_test_processed= preprocessor.transform(X_test)
```
